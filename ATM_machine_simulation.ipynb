```python
class Account:

    def __init__(self, user_id, pin, balance=0, transactions = None):
        self.user_id = user_id
        self.pin = pin
        self.balance = balance
        self.transactions = transactions if transactions is not None else []

    def check_pin(self, pin):
        return self.pin == pin

    def deposit(self, amount):
        self.balance += amount
        self.transactions.append(('+', amount))
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            return 'Insufficient balance'
        else:
            self.balance -= amount
            self.transactions.append(('-', amount))
            return self.balance

    def get_transaction_history(self):
        return self.transactions if self.transactions else 'No transactions available'

    def transfer(self, target_account, amount):
        if amount > self.balance:
            return 'Insufficient balance'
        else:
            self.withdraw(amount)
            target_account.deposit(amount)
            return 'amount transferd /n currunt balance:', self.balance
    def check_bal(self):
      return self.balance

class ATM:
    def __init__(self):
        self.accounts = {}
    def create_account(self, x, y, z=0):
        self.accounts[x] = Account(x, y, z
                                   )

    def access_account(self):
        x = input("Enter your user ID (use small letters only): ")
        y = input("Enter your pin: ")

        if x in self.accounts and self.accounts[x].check_pin(y):
            return self.accounts[x]
        else:
            return None

    def run(self):
        while True:
            print("\nATM Interface")
            print("*******************************************")
            print("1. Access Account")
            print("2. Quit")
            print("*******************************************")
            print("*******************************************")
            choice = input("Choose an option: ")
            if choice == '1':
                account = self.access_account()
                if account:
                    while True:
                        print("\n1. Transaction History")
                        print("2. Withdraw")
                        print("3. Deposit")
                        print("4. Transfer")
                        print("5. Check balance")
                        print("6. Quit")
                        print("*******************************************")
                        operation = input("Choose an operation: ")
                        if operation == '1':
                          print("*******************************************")
                          print("Transaction history", account.get_transaction_history())
                          print("*******************************************")
                        elif operation == '2':
                          amount = float(input("Enter amount to withdraw: "))
                          print("*******************************************")
                          print("Amount withdrawn successfully \n your currunt balance",account.withdraw(amount))
                          print("*******************************************")
                        elif operation == '3':
                            amount = float(input("Enter amount to deposit:"))
                            print("*******************************************")
                            print("Amount deposited successfully \n Your currunt Balance:", account.deposit(amount))
                            print("*******************************************")
                        elif operation == '4':
                            target_id = input("Enter target user ID:")
                            amount = float(input("Enter amount to transfer:"))
                            if target_id in self.accounts:
                                print("*******************************************")
                                print(account.transfer(self.accounts[target_id], amount))
                                print("*******************************************")
                            else:
                              print("*******************************************")
                              print("Target account not found.")
                              print("*******************************************")
                        elif operation == '5':
                          print("*******************************************")
                          print("your balance is:",account.check_bal())
                          print("*******************************************")
                        elif operation == '6':
                          print("XXXXXXXXXXXXXXXXXXXXXXXXXXXX")
                          print("Thank you! PRESS 2 TO EXIT")
                          print("XXXXXXXXXXXXXXXXXXXXXXXXXXXX")
                          break
                        else:
                          print("*******************************************")
                          print("ERROR! : you had enter Invalid operation. choose correct option")
                          print("*******************************************")
                else:
                  print("*******************************************")
                  print("ERROR!: Invalid user ID or pin. please enter correct user id & pin")
                  print("******************************************")
            elif choice == '2':
              print("*******************************************")
              print("SESSION TERMINATED! HAVE NICE DAY THANK YOU ;)")
              print("*****************************************User2**")              break
            else:
              print("*******************************************")
              print("ERROR!: Invalid choice. Please enter 1 or 2")
              print("*******************************************")

# Create an ATM inst
atm = ATM()
atm.create_account('user1', '1234', 1000)
atm.create_account('user2', '5678', 500)

# Run the ATM interface
atm.run()
```
